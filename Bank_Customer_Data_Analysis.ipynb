## Question 1

### Demographic Analysis

The first objective focuses on understanding the distribution of customers based on demographic factors and identifying differences in financial metrics across these groups.

```python
## Distribution of Customers by geography.

plt.figure(figsize=(8, 6) )
sb.countplot(data=df, x='Geography',
              palette='viridis')
plt.title('Distribution of Customers by Geography')
plt.xlabel('Geography')
plt.ylabel('Number of Customers')
plt.show()
```

```python
df.tail()
```

#### Observation:
France has higher customer base in the data.

```python
## Distribution of customers by Gender

plt.figure(figsize=(8, 6) )
sb.countplot(data=df, x='Gender', palette='Set2')
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()
```

#### Observation:

Male customers is higher than Female

```python
## Distribution by Age Group
## Definining age groups

bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(8, 6) )
sb.countplot(data=df, x='AgeGroup', order=labels, palette='plasma')
plt.title('Distribution of Customers by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.show()
```

## Observation

Customers within the Age Group range 35-44 has the highest accounts followed by Age Groups 25-34

```python
## Distribution by Geography and Gender

plt.figure(figsize=(10, 6))
sb.countplot(data=df, x='Geography', hue='Gender', palette={'Female': 'salmon', 'Male': 'lightblue'})
plt.title('Distribution of Customers by Geography and Gender')
plt.xlabel('Geography')
plt.ylabel('Number of Customers')
plt.legend(title='Gender')
plt.show()
```

## Observation

France has the highest customers with an account. Male customers has more accounts than the Female.

```python
## Age Groups
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

## Average Credit Score by Age Group
plt.figure(figsize=(10, 6))
avg_credit_by_age = df.groupby('AgeGroup')['CreditScore'].mean().reset_index()
sb.barplot(data=avg_credit_by_age, x='AgeGroup', y='CreditScore', order=labels, palette='coolwarm')
plt.title('Average Credit Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Credit Score')
plt.show()
```

## Observation

We can observe that the credit score slightly increase at age 65+

```python
## Average Credit Score by Gender
plt.figure(figsize=(6, 5))
avg_credit_by_gender = df.groupby('Gender')['CreditScore'].mean().reset_index()
sb.barplot(data=avg_credit_by_gender, x='Gender', y='CreditScore', palette='Set2')
plt.title('Average Credit Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Credit Score')
plt.show()
```

## Observation

Result shows that both Female and Male has an almost same credit score ratings.

```python
## Average Credit Score by Geography
plt.figure(figsize=(8, 6))
avg_credit_by_geography = df.groupby('Geography')['CreditScore'].mean().reset_index()
sb.barplot(data=avg_credit_by_geography, x='Geography', y='CreditScore', palette='viridis')
plt.title('Average Credit Score by Geography')
plt.xlabel('Geography')
plt.ylabel('Average Credit Score')
plt.show()
```

## Observation

Result shows that France, Germany and Spain all have almost same credit score rating.

```python
## Average Credit Score by Age Group and Gender
plt.figure(figsize=(12, 7))
sb.barplot(data=df, x='AgeGroup', y='CreditScore', hue='Gender', order=labels, palette={'Female': 'salmon', 'Male': 'lightblue'})
plt.title('Average Credit Score by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Average Credit Score')
plt.legend(title='Gender')
plt.show()
```

## Observations

--- It is observed that Male within the age range 18-24 and 55-64 has higher credit score rating while age range 65+ has almost same age range.

```python
## Average Credit Score by Gender and Geography
plt.figure(figsize=(10, 6))
sb.barplot(data=df, x='Gender', y='CreditScore', hue='Geography', palette='Spectral')
plt.title('Average Credit Score by Gender and Geography')
plt.xlabel('Gender')
plt.ylabel('Average Credit Score')
plt.legend(title='Geography')
plt.show()
```

```python
## Average Credit Score by Age Group and Geography
plt.figure(figsize=(14, 8))
sb.barplot(data=df, x='Geography', y='CreditScore', hue='AgeGroup', order=['France', 'Spain', 'Germany'], hue_order=labels, palette='coolwarm')
plt.title('Average Credit Score by Geography and Age Group')
plt.xlabel('Geography')
plt.ylabel('Average Credit Score')
plt.legend(title='Age Group')
plt.show()
```

## Question 2

### Customer Engagement Patterns

--- This objective is about investigating the relationship between tenure, engagement metrics and customer behaviour.

```python
correlation = df['Tenure_x'].corr(df['NumOfProducts'])
print(f"The correlation between Customer Tenure and Number of Products is: {correlation:.2f}")

plt.figure(figsize=(8, 6))
sb.barplot(data=df, x='Tenure_x', y='NumOfProducts')
plt.title('Scatter Plot of Customer Tenure vs. Number of Products')
plt.xlabel('Customer Tenure (Years)')
plt.ylabel('Number of Products')
plt.grid(True)
plt.show()
```

```python
## analyzing relationship between tenure and number of products
correlation_tenure_products = df['Tenure_x'].corr(df['NumOfProducts'])
print(f"Correlation between Tenure and Number of Products: {correlation_tenure_products:.2f}")

plt.figure(figsize=(8, 6))
sb.barplot(data=df, x='Tenure_x', y='NumOfProducts')
plt.title('Tenure vs. Number of Products')
plt.xlabel('Tenure (Years)')
plt.ylabel('Number of Products')
plt.grid(True)
plt.show()
```

## Question 3

### Financial Performance

Objective focuses on credit score distribution and balance / salary patterns across demographic groups

```python
## Age Groups
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 6))
sb.histplot(df['CreditScore'], bins=30, kde=True, color='skyblue')
plt.title('Overall Distribution of Credit Scores')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()
```

```python
## Credit Score Distribution by Age Group
plt.figure(figsize=(8, 6))
sb.barplot(data=df, x='Age', y='CreditScore', order='labels', palette='coolwarm')
plt.title('Credit Score Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Credit Score')
plt.show()
```

```python
# --- 3. Credit Score Distribution by Gender ---
plt.figure(figsize=(8, 6) )
sb.barplot(data=df, x='Gender', y='CreditScore', palette='Set2')
plt.title('Credit Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Credit Score')
plt.show()
```

```python
# --- 4. Credit Score Distribution by Geography ---
plt.figure(figsize=(10, 6))
sb.barplot(data=df, x='Geography', y='CreditScore', palette='viridis')
plt.title('Credit Score Distribution by Geography')
plt.xlabel('Geography')
plt.ylabel('Credit Score')
plt.show()
```

```python
# Question 4
### Objective involves comparing the financial and demographic characteristics of customers who exited versus those who stayed.

# --- Age Distribution by Exited Status ---
df['Exited'] = df['Exited'].map({0: 'Remained', 1: 'Exited'})


plt.figure(figsize=(10, 6))
sb.histplot(data=df, x='Age', hue='Exited', multiple='stack', palette={'Remained': 'skyblue', 'Exited': 'salmon'}, bins=30)
plt.title('Age Distribution by Customer Exited Status')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.legend(title='Exited')
plt.show()
```

```python
# --- Gender Distribution by Exited Status ---
plt.figure(figsize=(6, 5))
sb.barplot(data=df, x='Gender', hue='Exited', palette={'Remained': 'skyblue', 'Exited': 'salmon'})
plt.title('Gender Distribution by Customer Exited Status')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.legend(title='Exited')
plt.show()
```

```python
# --- Geography Distribution by Exited Status ---
plt.figure(figsize=(8, 6))
sb.barplot(data=df, x='Geography', hue='Exited', palette={'Remained': 'skyblue', 'Exited': 'salmon'})
plt.title('Geography Distribution by Customer Exited Status')
plt.xlabel('Geography')
plt.ylabel('Number of Customers')
plt.legend(title='Exited')
plt.show()
```
